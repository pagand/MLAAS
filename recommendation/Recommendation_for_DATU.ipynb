#First recomendation system:
simple based on user ranks


```python
import pandas as pd

# Load Movies Metadata
metadata = pd.read_csv('data_induced.csv', low_memory=False)

# Print the first three rows
metadata.head(20)
```

```python
# Calculate mean of vote average column
C = metadata['vote_average'].mean()
print(C)
```

```python
# Calculate the minimum number of votes required to be in the chart, m
m = metadata['vote_count'].quantile(0.1)
print(m)
```

```python
# Filter out all qualified library into a new DataFrame
q_library = metadata.copy().loc[metadata['vote_count'] >= m]
q_library.shape
```

```python
metadata.shape
```

```python
# Function that computes the weighted rating of each movie
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
# Define a new feature 'score' and calculate its value with `weighted_rating()`
q_library['score'] = q_library.apply(weighted_rating, axis=1)
```

```python
#Sort movies based on score calculated above
q_library = q_library.sort_values('score', ascending=False)

#Print the top 15 movies
q_library[['title', 'vote_count', 'vote_average', 'score']].head(10)
```

```python
def dataframe_related(df, category):
  return df.loc[df['category'] == category]
```

```python
categories = ["Introduction","data", "Supervised learning","optimization","Algorithm"]
category = categories[4]
df = dataframe_related(q_library, category)
q_df = df.sort_values('score', ascending=False)
q_df[['title', 'vote_count', 'vote_average', 'score']].head(3)
```

#Second recommendation:
Content-Based Recommender

```python
# Import Pandas
import pandas as pd

# Load Movies Metadata
metadata = pd.read_csv('data_induced.csv', low_memory=False)

# Print the first three rows
metadata.head(8)
```

```python
#Print plot overviews of the first 5 movies.
metadata['overview'].head()
```

```python
#Import TfIdfVectorizer from scikit-learn
from sklearn.feature_extraction.text import TfidfVectorizer

#Define a TF-IDF Vectorizer Object. Remove all english stop words such as 'the', 'a'
tfidf = TfidfVectorizer(stop_words='english')

#Replace NaN with an empty string
metadata['overview'] = metadata['overview'].fillna('')

#Construct the required TF-IDF matrix by fitting and transforming the data
tfidf_matrix = tfidf.fit_transform(metadata['overview'])

#Output the shape of tfidf_matrix
tfidf_matrix.shape
```

```python
vocab = tfidf.vocabulary_
print(vocab)
```

```python
tfidf_matrix
```

```python
#Array mapping from feature integer indices to feature name.
tfidf.get_feature_names_out()[1:10]
```

Let's use cosine similarity

![Alt text](https://res.cloudinary.com/dyd911kmh/image/upload/f_auto,q_auto:best/v1590782185/cos_aalkpq.png "cosine similarity")

```python
# Import linear_kernel
from sklearn.metrics.pairwise import linear_kernel

# Compute the cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
cosine_sim.shape
```

```python
cosine_sim[0]
```

```python
#Construct a reverse map of indices and movie titles
indices = pd.Series(metadata.index, index=metadata['title']).drop_duplicates()
```

```python
indices[:10]
```

```python
# Function that takes in movie title as input and outputs most similar library
def get_recommendations(title, num_recom, cosine_sim=cosine_sim):
    # Get the index of the movie that matches the title
    idx = indices[title]

    # Get the pairwsie similarity scores of all library with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    metadata['scores'] = cosine_sim[idx]

    # Sort the library based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the num_recom most similar library
    sim_scores = sim_scores[0:num_recom]

    # Get the movie indices
    lib_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar library
    return metadata[['number', 'title','category','scores']].iloc[lib_indices]
```

```python
# approach 2-1: based on the overview similiarity of the input title
categories = ["Introduction","data", "Supervised learning","optimization","Algorithm"]
get_recommendations(title = 'machine learning', num_recom = 10)
```

#Third recommendation:
Hybrid Recommender

```python
def get_similar_keywords(keywords,num_recom ):
  seri = metadata['keywords'].str.split(',')
  seri = seri.apply(lambda x: [j.strip().lower() for j in x])
  scores = [(index,len(x)+ len(keywords) - len(set(keywords+x)))  for index,x in enumerate(seri)]
  scores = sorted(scores, key=lambda x: x[1], reverse=True)

  metadata['scores'] = seri.apply(lambda x: (len(x)+ len(keywords) - len(set(keywords+x)))/len(x))


  # Get the scores of the 3 most similar library
  scores = scores[0:num_recom]

  # Get the movie indices
  lib_indices = [i[0] for i in scores]

  # Return the top 10 most similar library
  return metadata[['number', 'title','category','scores']].iloc[lib_indices]
```

```python
# approach 2-2: based on the similarities in keywords

list_keyword = ['machine learning', 'languages', 'pandas', 'numpy', 'training', 'libraries','study', 'library', 'program','multi class' ,'good']
get_similar_keywords(keywords = list_keyword, num_recom = 10)
```

```python
def get_similar_new_keywords(keyword, num_recom):
  vectorizer = TfidfVectorizer(vocabulary=vocab, stop_words='english')
  titles_tfidf = vectorizer.fit_transform(keyword)
  cosine_sim1 = linear_kernel(titles_tfidf, tfidf_matrix)
  # total_sum = [(indx ,sum_item) for indx, sum_item in enumerate(sum(cosine_sim1))]
  # scores = sorted(total_sum, key=lambda x: x[1], reverse=True)
  seri = pd.Series(sum(cosine_sim1))
  lib_indices = seri.sort_values(ascending=False)[:num_recom].index
  metadata['scores'] = seri

  return metadata[['number', 'title','category','scores']].iloc[lib_indices]
```
